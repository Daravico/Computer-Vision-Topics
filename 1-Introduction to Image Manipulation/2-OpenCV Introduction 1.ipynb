#### BASIC IMPORT WITH DETAILS CHECKING

```python
# Basic Importation.
%matplotlib inline

import cv2
import matplotlib.pyplot as plt

# Reading the image.
pic_path = "img/starry_night.jpg"
img = cv2.imread(pic_path)

# Showing the type of the image.
print(type(img))

# Changing the color (OpenCV works with BGR, Matplotlib with RGB)
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Showing the image.
plt.imshow(RGB_img)
plt.show()

# Showing the size and a pixel.
print(RGB_img.shape)
print(RGB_img[0][0])
```

#### GRAY IMAGE AND CMAPS

```python
# CMAPS.
import cv2
import matplotlib.pyplot as plt

# Reading the image again.
pic_path = "img/starry_night.jpg"
img = cv2.imread(pic_path)

# Grayscale. Showing that it requires the cmap again.
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
plt.imshow(gray_img, cmap='magma')
plt.imshow(gray_img, cmap='bone')
plt.imshow(gray_img, cmap='pink')
plt.imshow(gray_img, cmap='cool')
plt.imshow(gray_img, cmap='afmhot')
plt.imshow(gray_img, cmap='BrBG')
plt.imshow(gray_img, cmap='Spectral')
plt.imshow(gray_img, cmap='coolwarm')
plt.show()

print(gray_img.shape)
```

#### RESIZING

```python
# RESIZING.

import cv2
import matplotlib.pyplot as plt

pic_path = "img/beach.jpg"
img = cv2.imread(pic_path)

# plt.imshow(img)

# Requires recoloring.
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Direct resizing (Pixel-wise).
resized = cv2.resize(RGB_img, (1000, 500))
# resized = cv2.resize(RGB_img, (500, 1000))

# Resizing scale.
w_ratio = 0.5
h_ratio = 1
resized = cv2.resize(RGB_img, (0,0), RGB_img, w_ratio, h_ratio)

plt.imshow(resized)
plt.show()
```

#### FLIP

```python
# Used in DL to generate more datasets (cases).
import cv2
import matplotlib.pyplot as plt

pic_path = "img/beach.jpg"
img = cv2.imread(pic_path)

RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

flipped = cv2.flip(RGB_img, 0)
flipped = cv2.flip(RGB_img, 1)
flipped = cv2.flip(RGB_img, -1)
flipped = cv2.flip(RGB_img, 2)


plt.imshow(flipped)
plt.show()
```

#### SAVE IMG
