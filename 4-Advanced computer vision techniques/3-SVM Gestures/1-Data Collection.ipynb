```python
import cv2

cap = cv2.VideoCapture(0)

# Variables.
count = 0
limit = 300
id_gesture = 0

# Gestures that will be captured.
gestures = {
    'empty': 0,
    'one': 0,
    "two": 0,
    'palm': 0,
}

# Saving the details of the dictionary.
gesture_names = list(gestures.keys())
limit_gestures = len(gestures.keys())

# Variables for the region in the frame.
x = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH) // 2 - cap.get(cv2.CAP_PROP_FRAME_WIDTH) // 8)
y = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT) // 4)
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH) // 4)
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT) // 2)

# Main Loop.
while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    if not ret:
        break

    if id_gesture >= limit_gestures:
        break
    
    k = cv2.waitKey(10)
    
    # Pressing the letter 'q' to end the program. 
    if k == ord('q'):
        break

    if count >= limit:
        cv2.putText(frame, 
                f"Press 'n' to continue...", 
                (400, 150), 
                cv2.FONT_ITALIC, 
                0.5, (255,0,0), 1)
        cv2.rectangle(frame, (x,y), (x+w, y+h), (255,255,255), 2)

        if k == ord('n'):
            count = 0
            id_gesture+=1    
    
    
    elif k == ord('t'):
        count+=1
        cv2.imwrite(f'data/Gesture_{str(id_gesture)}_{str(count)}.jpg', gray[y:y+h, x:x+w])
        cv2.rectangle(frame, (x,y), (x+w, y+h), (0,0,255), 2)
    
    else:
        cv2.rectangle(frame, (x,y), (x+w, y+h), (0,255,0), 2)

    

    cv2.putText(frame, 
                f"Gesture: {gesture_names[id_gesture]}", 
                (400, 50), 
                cv2.FONT_ITALIC, 
                0.75, (255,0,0), 1)
    
    cv2.putText(frame, 
                f"Count: {count}", 
                (400, 100), 
                cv2.FONT_ITALIC, 
                0.75, (255,0,0), 1)

    cv2.imshow("Frame", frame)    

cap.release()
cv2.destroyAllWindows()
```
