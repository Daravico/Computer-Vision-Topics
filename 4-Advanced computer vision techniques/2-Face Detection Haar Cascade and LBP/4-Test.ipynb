```python
import cv2

cap = cv2.VideoCapture(0)

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer.read("FaceModel.yml")

names = {
    1: "David",
    2: "Random"
}

while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, 1.3, 5)


    if len(faces) > 0:
        largest_face = max(faces, key=lambda rectangle: rectangle[2] * rectangle[3])
        x, y, w, h = largest_face

        label, confidence = recognizer.predict(gray[y:y+h, x:x+w])

        print(label)
        print(confidence)
        
        if confidence < 60:
            cv2.putText(frame, names[label], (x + w // 3, y - 30), cv2.FONT_ITALIC, 1, (255,0,0), 1)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
        else:
            cv2.putText(frame, "UNKNOWN", (x + w // 3, y - 30), cv2.FONT_ITALIC, 1, (0,0,255), 1)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)

    cv2.imshow("Frame", frame)

    k = cv2.waitKey(1)

    if k == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
