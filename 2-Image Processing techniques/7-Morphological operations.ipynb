# Erosion

```python
# EROSION
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_path = "img/circle.png"
img = cv2.imread(img_path).astype(np.float32) / 255
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel = np.ones((5,5), dtype=np.uint8)
img_morph = cv2.erode(img, kernel, iterations=7)

plt.imshow(img)
plt.figure()
plt.imshow(img_morph)
```

## Opening

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_path = "img/circle.png"
img = cv2.imread(img_path).astype(np.float32) / 255
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap="gray")
plt.figure()

kernel = np.ones((5,5), dtype=np.uint8)

white_noise = 255 * (np.random.randint(low=0, high=2, size=(img.shape[0], img.shape[1])))
plt.figure()
plt.imshow(white_noise, cmap='gray')

noisy_img = img + white_noise

alpha = 0.7  
beta = 0.9
noisy_img = cv2.addWeighted(img, alpha, white_noise.astype(np.float32) / 255 , beta, 0)

plt.figure()
plt.imshow(noisy_img, cmap='gray')

opening = cv2.morphologyEx(noisy_img, cv2.MORPH_OPEN, kernel)
plt.figure()
plt.imshow(opening, cmap="gray")
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Leer la imagen en escala de grises.
img_path = "img/circle.png"
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) / 255.0

plt.imshow(img, cmap="gray")
plt.title("Imagen Original en Grises")
plt.axis('off')
plt.figure()

# Aadir "salt" (ruido blanco) solo a regiones negras.
salt_noise = np.random.randint(0, 2, size=img.shape)
noisy_img = img.copy()
noisy_img[(img == 0) & (salt_noise == 1)] = 1  # Aadir "salt" solo a pxeles negros.

plt.imshow(noisy_img, cmap='gray')
plt.title("Imagen con Ruido Salt en Regiones Negras")
plt.axis('off')
plt.figure()

# Corregir la imagen con una operacin morfolgica de apertura (opening).
kernel = np.ones((3, 3), dtype=np.uint8)
opening = cv2.morphologyEx(noisy_img, cv2.MORPH_OPEN, kernel)

plt.imshow(opening, cmap="gray")
plt.title("Imagen Corregida con Opening")
plt.axis('off')
plt.show()
```

# Closing

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Leer la imagen en escala de grises.
img_path = "img/circle.png"
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) / 255.0

plt.imshow(img, cmap="gray")
plt.title("Imagen Original en Grises")
plt.axis('off')
plt.figure()

# Aadir ruido blanco (pepper) a toda la imagen.
white_noise = np.random.randint(0, 2, size=img.shape)
noisy_img = img.copy()
noisy_img[white_noise == 1] = 1  # Aadir ruido blanco a los pxeles seleccionados aleatoriamente.

plt.imshow(noisy_img, cmap='gray')
plt.title("Imagen con Ruido Blanco")
plt.axis('off')
plt.figure()

# Corregir la imagen con una operacin morfolgica de cierre (closing).
kernel = np.ones((5,5), dtype=np.uint8)

img = cv2.dilate(img, (5,5), iterations=35)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.imshow(closing, cmap="gray")
plt.title("Imagen Corregida con Closing")
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_path = "img/circle_black.png"
img = cv2.imread(img_path).astype(np.float32) / 255
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap="gray")
plt.title("Imagen Original")
plt.axis('off')
plt.figure()

# Generar ruido negro (pxeles negros).
black_noise = np.random.randint(low=0, high=2, size=(img.shape[0], img.shape[1])).astype(np.float32)
black_noise = black_noise * -1  # Multiplica por -1 para crear ruido negativo.

# Aplicar el ruido negro a la imagen original.
noisy_img = img + black_noise
noisy_img = np.clip(noisy_img, 0, 1)  # Asegurar que la imagen resultante est en el rango [0, 1]

plt.imshow(noisy_img, cmap='gray')
plt.title("Imagen con Ruido Negro")
plt.axis('off')
plt.figure()

# Aplicar operacin morfolgica de apertura (opening).
kernel = np.ones((5,5), dtype=np.uint8)
opening = cv2.morphologyEx(noisy_img, cv2.MORPH_OPEN, kernel)
plt.imshow(opening, cmap="gray")
plt.title("Imagen despus de Opening")
plt.axis('off')

plt.show()
```
