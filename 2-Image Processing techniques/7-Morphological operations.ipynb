# Morphological Operation

# Erosion

```python
# EROSION
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_path = "img/circle.png"
img = cv2.imread(img_path).astype(np.float32) / 255
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel = np.ones((5,5), dtype=np.uint8)
img_morph = cv2.erode(img, kernel, iterations=7)

plt.imshow(img)
plt.figure()
plt.imshow(img_morph)
```

## Dilatation

```python
# EROSION
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_path = "img/circle.png"
img = cv2.imread(img_path).astype(np.float32) / 255
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel = np.ones((5,5), dtype=np.uint8)
img_morph = cv2.dilate(img, kernel, iterations=7)

plt.imshow(img)
plt.figure()
plt.imshow(img_morph)
```

## Opening

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_path = "img/circle.png"
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) / 255.0  # Normalizar a rango 0-1

plt.imshow(img, cmap='gray')
plt.title("Imagen Original")
plt.axis('off')
plt.figure()

# Probability matrix and applied mask.
probability_matrix = np.random.rand(img.shape[0], img.shape[1])
modify_mask = (img == 0) & (probability_matrix <= 0.3)
img[modify_mask] = 1

plt.imshow(img, cmap='gray')
plt.title("Imagen Después de la Modificación Bitwise (Añadir Blanco)")
plt.axis('off')
plt.show()

# Opening operation.
kernel = np.ones((3, 3), dtype=np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

plt.imshow(opening, cmap="gray")
plt.title("Opening correction")
plt.axis('off')
plt.show()
```

# Closing

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_path = "img/circle.png"
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) / 255.0  

plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.figure()

# Probability matrix and applied mask.
probability_matrix = np.random.rand(img.shape[0], img.shape[1])
modify_mask = (img == 1) & (probability_matrix <= 0.5)
img[modify_mask] = 0

plt.imshow(img, cmap='gray')
plt.title("Image with black noise")
plt.axis('off')
plt.show()

# Closing operation.
kernel = np.ones((3, 3), dtype=np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.imshow(closing, cmap="gray")
plt.title("Closing correction")
plt.axis('off')
plt.show()
```
